```python
categories_keywords = {
    "Princesses/Royalty": ["Princess", "Royalty", "Castle", "Crown"],
    "Fairies/Fantasy": ["Fairytopia", "Magic", "Rainbow", "Mariposa", "Mermaid", "Pegasus", "Nutcracker", "Rapunzel", "Swan Lake", "Thumbelina", "Diamond Castle", "Secret Door", "Pearl Princess", "12 Dancing Princesses", "Fairy"],
    "Adventure/Modern": ["Spy Squad", "Star Light Adventure", "Dolphin Magic", "Princess Adventure", "Lost Birthday", "Big City, Big Dreams", "Epic Road Trip", "Skipper", "Babysitting Adventure", "Mermaid Power", "It Takes Two", "Touch of Magic", "The Movie", "Pony Tale", "Puppy Adventure", "Puppy Chase", "Pink Shoes", "Popstar", "Fashion Fairytale", "Rock 'N Royals", "Island Princess", "Diaries"],
    "Classic Tales": ["Nutcracker", "Rapunzel", "Swan Lake", "Princess and the Pauper", "Christmas Carol", "Thumbelina", "Three Musketeers"],
    "Musicals/Dance": ["Popstar", "Pink Shoes", "Rock 'N Royals", "12 Dancing Princesses", "Singing", "Dance"]
}

# Initialize the classified_movies dictionary with empty lists for each category
classified_movies = {category: [] for category in categories_keywords.keys()}
classified_movies["Other"] = [] # Add an 'Other' category for movies not fitting specific criteria

# Iterate through each movie and classify it
for movie in barbie_movies:
    classified_into_any_category = False
    for category, keywords in categories_keywords.items():
        for keyword in keywords:
            if keyword.lower() in movie.lower():
                # Avoid duplicate entries if a movie fits multiple keywords within the same category
                if movie not in classified_movies[category]:
                    classified_movies[category].append(movie)
                classified_into_any_category = True

    # If the movie did not fit into any specific category, add it to 'Other'
    if not classified_into_any_category:
        classified_movies["Other"].append(movie)

# Display the classified movies for verification
print("--- Classified Barbie Movies ---")
for category, movies in classified_movies.items():
    if movies:
        print(f"\n{category} ({len(movies)} movies):\n  " + "\n  ".join(movies))
    else:
        print(f"\n{category}: No movies classified")
```
