Le cryptage de César (empereur romain) est le tout premier code de cryptage qui ait existé.

La méthode est simple: il suffit de décaler toutes les lettres de l'alphabet du même nombre de lettres.
Par exemple, en choisissant un décalage de 3, le a devient le d, le b devient le e, ....,le w devient z
Pour la fin de l'alphabet, il suffit de revenir au début : le x devient a, le y devient b et le x devient
c.
