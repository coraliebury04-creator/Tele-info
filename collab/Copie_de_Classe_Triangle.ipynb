### A - Coder une classe **Triangle** contenant 3 points **A**, **B** et **C**, un **constructeur** qui initialise ce triangle avec 3 points de coordonnées (0, 0), une méthode **perimetre()** qui renvoie la somme des longueurs des segments AB, BC et CA, et une méthode **barycentre()** qui renvoie un Point **G** avec les coordonnées du centre de gravité du triangle ABC.

image.png

```python
import math

class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def distance(self, autre_point):
        return math.sqrt((self.x - autre_point.x)**2 + (self.y - autre_point.y)**2)

    def __repr__(self):
        return f"({self.x:.2f}, {self.y:.2f})"

class Triangle:
    def __init__(self):
        # Initialise le triangle avec 3 points de coordonnées (0, 0)
        self.A = Point(0, 0)
        self.B = Point(0, 0)
        self.C = Point(0, 0)

    def perimetre(self):
        # Renvoie la somme des longueurs des segments AB, BC et CA
        ab = self.A.distance(self.B)
        bc = self.B.distance(self.C)
        ca = self.C.distance(self.A)
        return ab + bc + ca

    def barycentre(self):
        # Renvoie un Point G avec les coordonnées du centre de gravité
        xG = (self.A.x + self.B.x + self.C.x) / 3
        yG = (self.A.y + self.B.y + self.C.y) / 3
        return Point(xG, yG)
```

### B - Tester la classe **Triangle** en créant 3 points **A**, **B** et **C**, puis en déplaçant A vers **(3, 2)**, B vers **(7, 5)**, et C vers **(5, 8)**, puis en affichant le  périmetre du triangle **ABC** et les coordonnées du point **G**, centre de gravité de ce triangle.

```python
# Create Point objects for A, B, and C with the specified coordinates
pointA = Point(3, 2)
pointB = Point(7, 5)
pointC = Point(5, 8)

triangle_test = Triangle()

triangle_test.A = pointA
triangle_test.B = pointB
triangle_test.C = pointC

perimeter = triangle_test.perimetre()
print(f"Le périmètre du triangle ABC est : {perimeter:.2f}")

barycenter_G = triangle_test.barycentre()
print(f"Les coordonnées du barycentre G sont : ({barycenter_G.x:.2f}, {barycenter_G.y:.2f})")
```

### C - Afficher le triangle ABC et son centre de gravité dans un graphe en utilisant la bibliothèque MatPlotLib


```python
import matplotlib.pyplot as plt

# --- Utilisation des classes Point et Triangle définies précédemment ---
pA = Point(3, 2)
pB = Point(7, 5)
pC = Point(5, 8)

triangle = Triangle() # Create a Triangle instance using the no-argument constructor
triangle.A = pA
triangle.B = pB
triangle.C = pC

g = triangle.barycentre()

# --- Tracé ---
plt.figure(figsize=(8, 8)) # Increased figure size for better readability

# Plotting the triangle
xs, ys = [pA.x, pB.x, pC.x, pA.x], [pA.y, pB.y, pC.y, pA.y]
plt.plot(xs, ys, 'b-', label='Triangle ABC')

# Plotting the vertices A, B, C
plt.scatter([pA.x, pB.x, pC.x], [pA.y, pB.y, pC.y], c=['red', 'green', 'cyan'], zorder=5)

# Adding labels for vertices A, B, C
plt.text(pA.x - 0.2, pA.y - 0.5, 'A', color='red', fontsize=12, ha='right')
plt.text(pB.x + 0.2, pB.y - 0.5, 'B', color='green', fontsize=12, ha='left')
plt.text(pC.x + 0.2, pC.y + 0.3, 'C', color='cyan', fontsize=12, ha='left')

# Calculating midpoints for medians
m_ab = Point((pA.x + pB.x) / 2, (pA.y + pB.y) / 2)
m_bc = Point((pB.x + pC.x) / 2, (pB.y + pC.y) / 2)
m_ca = Point((pC.x + pA.x) / 2, (pC.y + pA.y) / 2)

# Plotting medians
plt.plot([pA.x, m_bc.x], [pA.y, m_bc.y], 'r--', label='Médianes')
plt.plot([pB.x, m_ca.x], [pB.y, m_ca.y], 'r--')
plt.plot([pC.x, m_ab.x], [pC.y, m_ab.y], 'r--')

# Plotting the barycenter G
plt.plot(g.x, g.y, 'kx', markersize=12, markeredgewidth=3, label=f'G {g}')

plt.title('Triangle, Médianes et Centre de Gravité')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
