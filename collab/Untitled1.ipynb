```python
import math

# Lecture de l'entrée
N = int(input("Entrez un nombre entier: "))

# Initialisation
quotient = math.floor(N / 2)
reste = N % 2
resultat = []
indice = 0

# Conversion en binaire
while quotient != 0:
    resultat.append(reste)
    reste = quotient % 2  # Corrected assignment
    quotient = math.floor(quotient / 2)
    indice += 1

# Dernier bit
resultat.append(1)

# Affichage du résultat en inversant l'ordre
while indice >= 0:
    print(resultat[indice], end="")
    indice -= 1

# Optionnel saut de ligne
print()
```
