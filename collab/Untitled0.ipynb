```python
total_notes = 0
nombre_notes = 5

for i in range(nombre_notes):
    while True:
        try:
            note_input = input(f"Entrez la note {i + 1} (sur 20) : ")
            note = float(note_input)
            if 0 <= note <= 20:
                total_notes += note
                break
            else:
                print("Veuillez entrer une note entre 0 et 20.")
        except ValueError:
            print("Veuillez entrer un nombre valide.")

moyenne = total_notes / nombre_notes

if moyenne < 10:
    print("Échec")
else:
    print("Réussi")

print(f"La moyenne est de : {moyenne:.2f}")
```

```python
while True:
    try:
        n_input = input("Entrez un entier pour afficher sa table de multiplication : ")
        n = int(n_input)
        break
    except ValueError:
        print("Veuillez entrer un entier valide.")

print(f"Table de multiplication de {n} :")
for i in range(1, 11):
    print(f"{n} × {i} = {n * i}")
```

```python
while True:
    try:
        day_number_input = input("Entrez un nombre entre 1 et 7 pour afficher le jour correspondant : ")
        day_number = int(day_number_input)

        days_of_week = {
            1: "Lundi",
            2: "Mardi",
            3: "Mercredi",
            4: "Jeudi",
            5: "Vendredi",
            6: "Samedi",
            7: "Dimanche"
        }

        if day_number in days_of_week:
            print(days_of_week[day_number])
            break
        else:
            print("numéro invalide")

    except ValueError:
        print("Veuillez entrer un nombre valide.")
```

```python
while True:
    try:
        number_input = input("Entrez un nombre réel : ")
        number = float(number_input)

        if number < 0:
            print("Négatif")
        elif number == 0:
            print("Nul")
        else:
            print("Positif")
        break
    except ValueError:
        print("Veuillez entrer un nombre valide.")
```

```python
while True:
    try:
        age_input = input("Entrez l'âge de la personne : ")
        age = int(age_input)

        if age < 12:
            print("Enfant")
        elif 12 <= age < 18:
            print("Adolescent")
        elif 18 <= age < 60:
            print("Adulte")
        else:
            print("Senior")
        break
    except ValueError:
        print("Veuillez entrer un âge valide.")
```
