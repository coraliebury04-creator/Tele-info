### Coder un programme permettant d'afficher :
- Les coordonnées GPS,
- La superficie,
- Le nombre d'habitants,

### d'un pays dont le nom est entré au clavier par l'utilisateur.
##### L'API permettant de récupérer ces données est ici :
https://restcountries.com/v3.1/name/{country_name}

#####Par exemple si on entre 'Martinique', l'URL sera :
https://restcountries.com/v3.1/name/Martinique



```python
import requests

country_name = input("Entrer le nom du pays : ")
api_url = f"https://restcountries.com/v3.1/name/{country_name}"

try:
    response = requests.get(api_url)
    response.raise_for_status()
    data = response.json()

    if data:
        country_data = data[0]
        coordinates = country_data.get('latlng', ['', ''])
        latitude = coordinates[0]
        longitude = coordinates[1]
        area = country_data.get('area', '')
        population = country_data.get('population', '')

        print(f"Latitude : {latitude} Longitude : {longitude}")
        print(f"Superficie : {area} km2")
        print(f"Habitants : {population}")
    else:
        print(f"Aucune donnée trouvée pour le pays : {country_name}")


except requests.exceptions.RequestException as e:
    print(f"Erreur lors de la requête : {e}")
```
